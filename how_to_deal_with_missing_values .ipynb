```python
import pandas as pd
# reading information from a CSV file
cars_df = pd.read_csv("cars_multi.csv", na_values = '?')
```

```python
# Returns a DataFrame where each cell is either True or False depending on that cell's null status.
cars_df.isnull()
```

```python
# To count the number of nulls in each column we use an aggregate function for summing:
cars_df.isnull().sum()
```

```python
# Now there is a row that could be removed. The method dropna will return a new dataframe
temp = cars_df.dropna()
temp.isnull().sum()
```

```python
# Lets put a new car into the dataframe for the sake of the exercise
a_row = pd.Series([399, 31.0, 4, 119.0, None, 2720, 19.4, 82, 1, "Awesome Car"], index=cars_df.columns)
cars_df = cars_df.append(a_row, ignore_index=True)
cars_df.tail(5)
```

```python
# Getting the horsepower column
horsepower = cars_df['horsepower']
# Getting the horsepower mean
horsepower_mean = horsepower.mean()
# Replacing missing values in the series and also the dataframe where it came from with the option "inplace=True"
horsepower.fillna(horsepower_mean, inplace=True)
```
